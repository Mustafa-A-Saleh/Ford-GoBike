## Preliminary Wrangling

Ford GoBike is a bike-sharing system that allows people to use bikes as a transportation mean through different stations in San Francisco Bay Area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# Loading Ford GoBike dataset
go_bike = pd.read_csv('ford_go_bike.csv')
```

### Quick overview and descriptive statistics for the dataset

```python
go_bike.head()
```

```python
go_bike.shape
```

```python
go_bike.dtypes
```

```python
go_bike.describe()
```

```python
# Drop unnecessary columns: [start_station_id, start_station_latitude, start_station_longitude, end_station_id, end_station_latitude, end_station_longitude, bike_id]
go_bike.drop(columns= ['start_station_id','start_station_latitude', 'start_station_longitude', 'end_station_id', 'end_station_latitude', 'end_station_longitude', 'bike_id'], inplace= True)
```

```python
# Drop null values [almost 5% and will not affect our analysis dramatically]
print(go_bike.isnull().sum())
go_bike.dropna(inplace=True)
```

```python
# Check for duplicates
go_bike.duplicated().sum()
```

```python
# change member_birth_year to integer
go_bike['member_birth_year']= go_bike['member_birth_year'].astype(int)
```

```python
# Change duration to minutes instead of seconds and rename column to (duration_min)
go_bike['duration_sec'] = (go_bike['duration_sec'] / 60).astype(int)
go_bike.rename(columns={'duration_sec':'duration_min'}, inplace=True)
```

```python
# Add new column 'age' by subtracting member_birth_year from 2019
go_bike['age'] = 2019 - go_bike['member_birth_year']
```

```python
# make sure all of the previous changes took place
go_bike.head()
```

```python
print(go_bike.shape)
```

## Univariate Exploration

We will try to figure out how duration is distributed

```python
# Plotting the relation between number of the trips and duration
binsize= 40
bins = np.arange(0, go_bike['duration_min'].max()+binsize, binsize)

plt.figure(figsize=[8, 8])
plt.hist(data = go_bike, x = 'duration_min', bins = bins);
plt.xlabel('Duration (min)')
plt.ylabel('Number of Trips')
plt.title('Number of Trips based on Duration');
```

```python
# To make it clear, we will use a log scale for the -duration plot
log_binsize = .09
bins = 10 ** np.arange(0, np.log10(go_bike['duration_min'].max())+log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = go_bike, x = 'duration_min', bins = bins)
plt.xscale('log')
ticks = [1, 2, 5, 10, 15, 20, 50, 100, 200, 500, 1400]
plt.xticks(ticks, ticks)
plt.xlabel('Duration (min)')
plt.ylabel('Number of Trips')
plt.title('Number of Trips based on Duration');

# Plotting vertical line showing the mean
plt.axvline(go_bike['duration_min'].mean(), color='r', lw=2);
```

It appears that most of our trips takes between 10 to 13 minutes with the mean around 12 minutes.

Now, we will see gender distribution

```python
# Plotting relation between Gender and Number of Users
plt.figure(figsize=[8,5])
sb.countplot(data = go_bike, x = 'member_gender',color = sb.color_palette()[0], order = go_bike['member_gender'].value_counts().index);
plt.xlabel('Gender')
plt.ylabel('Users')
plt.title('Gender Distribution among Users');
```

Previous plot shows that most of our users are male.

Let's see age distribution

```python
# Plotting relation Age and between Number of Trips 
plt.figure(figsize=[8,5])

plt.hist(data = go_bike, x = 'age', bins = 20);
plt.xlim([10,100]) #limiting age between 10 and 100 (more than 100 is not logic) 
plt.xlabel('Age')
plt.ylabel('Number of Users')
plt.title('Age Distribution');
```

Age of users increases and reach its peak around 30 years. After that duration is going down.

Now, with user type distribution

```python
# Showing percentage of subscribers and customers
plt.figure(figsize=[8,8])
plt.pie(go_bike['user_type'].value_counts(), labels=['Subscriber','Customer'], autopct='%1.0f%%', colors=['tab:blue', 'c'])
plt.title('User Type Distribution');
```

People who subscribed to our service equal 10 times regular customers

Let's see which are our top 10 start and end stations

```python
# creating subplot showing Top 10 start and end stations
fig, ax = plt.subplots(nrows=2, figsize = [8,8])

# Top 10 start stations
sb.countplot(data = go_bike, y = 'start_station_name', color = sb.color_palette()[0], 
        order = go_bike['start_station_name'].value_counts().index[0:10], ax = ax[0])

ax[0].set_ylabel('Top 10 Start Stations')
ax[0].set_xlabel('Number of Trips')

# Top 10 end stations
sb.countplot(data = go_bike, y = 'end_station_name', color = sb.color_palette()[0], 
        order = go_bike['end_station_name'].value_counts().index[0:10], ax = ax[1])

ax[1].set_ylabel('Top 10 end Stations')
ax[1].set_xlabel('Number of Trips');
```

'Market St at 10th St' and 'San Francisco Caltrain Station 2' are the most popular stations (as start and end stations) 

Here, we get to the effect of weekdays and weekends

```python
# changing start time type to datetime
go_bike.start_time = pd.to_datetime(go_bike.start_time)

# Add new column for days
go_bike['weekday'] = go_bike.start_time.dt.weekday

# Plotting number of starting trips feo each weekday
s= sb.countplot(data = go_bike, x = 'weekday', color = sb.color_palette()[0] )
s.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday'])
plt.ylabel('Number of Trips');
plt.xticks(rotation=45);
plt.title('Trips distribution through Weekdays');
```

The last plot indicates that trips in weekday are more than weekends.

## Bivariate Exploration

We start our bivariate exploration with the relation between duration and age

```python
# Plotting relation between Age and Trip Duration
plt.figure(figsize = [8, 6])
plt.scatter(data = go_bike, x = 'age', y = 'duration_min', alpha = .5)
plt.xlabel('Age (years)')
plt.ylabel('Duration (min)')
plt.title('Relation between Age and Trip Duration');
```

As age increases, trip duration decreases. Longer trips are related to people from 20 to 40, and most of trips last less than 30 minutes

Now, we move to the relation between gender and duration

```python
# This plot shows distribution of Duration based on Gender
plt.figure(figsize = [8, 5])
sb.boxplot(data = go_bike, y = 'member_gender', x = 'duration_min', color = sb.color_palette()[0])

plt.xlim([0, 40])
plt.xlabel('Duration (min)')
plt.ylabel('Gender')
plt.title('Distribution of Duration based on Gender');
```

Female users take a slight longer trips than other gender.

Let's see the relation between user type and duration

```python
plt.figure(figsize = [8, 5])
sb.boxplot(data = go_bike, y = 'user_type', x = 'duration_min', color = sb.color_palette()[0])
plt.xlim([0, 50])
plt.xlabel('Duration (min)')
plt.ylabel('User Type')
plt.title('Distribution of Duration based on User Type');
```

Customers take longer trips than Subscribers.

And here is the relation between weekday and trip duration

```python
plt.figure(figsize = [8, 5])
a= sb.boxplot(data = go_bike, x = 'weekday', y = 'duration_min', color = sb.color_palette()[0])
plt.ylim([0, 40])
a.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday'])
plt.xticks(rotation=45)
plt.ylabel('Duration (min)')
plt.xlabel('Weekday')
plt.title('Distribution of Duration based on Weekday');
```

## Multivariate Exploration

We will start with the relation between number of the trips throughout the week, associated with gender

```python
# Plotting number of starting trips for each weekday for different genders
s= sb.countplot(data = go_bike, x = 'weekday', hue='member_gender')
s.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday'])
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.title('Distribution of Trips Through weekdays based on Gender');
```

Male trips are obviously the dominant by 3 or 4 times through the week. 

To dig deeper, we will see the relation between duration, age and gender.

```python
# Plotting Trips Duration based on Age and Gender 
sb.lmplot(data= go_bike, x='age' , y="duration_min",  fit_reg=False, hue='member_gender', legend=False, scatter_kws={'alpha':0.5})
plt.legend(title ='Gender')

plt.xlabel('Age (years)')
plt.ylabel('Duration (min)')
plt.title('Trips Duration based on Age and Gender');
```

Through the years, trips duration decrease, but male users still have longer trips compared to female users.

We will see how  gender distributed regarding usre type

```python
# Plotting Users based on User Type and Gender 
g = sb.FacetGrid(data = go_bike, col = 'user_type', height = 5, aspect =1)
g.map_dataframe(sb.countplot, 'member_gender', alpha= None)
g.set_axis_labels('Gender', 'Number of Users')
g.fig.suptitle('Distribution of Users based on Gender and User Type', va= 'bottom');
```

For Subscribers, there is a huge gap between male (the majority of users) and female.
For customers, we don't see that gap.
